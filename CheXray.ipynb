##### Splitting the diffrent diseses classes

```python
# Atelectasis
atelectasis_indexes = []
atelectasis_images = []

# Cardiomegaly
cardiomegaly_indexes = []
cardiomegaly_images = []

# Effusion
effusion_indexes = []
effusion_images = []

# Infiltration
infiltration_indexes = []
infiltration_images = []

# Mass
mass_indexes = []
mass_images = []

# Nodule
nodule_indexes = []
nodule_images = []

# Pneumonia
pneumonia_indexes = []
pneumonia_images = []

# Pneumothorax
pneumothorax_indexes = []
pneumothorax_images = []
```

```python
cardiomegaly_indexes = []
cardiomegaly_images = []
for index, i in enumerate(data_entry.iloc[:, 1]):  
        
    # Atelectasis
    if 'Atelectasis' in i.split('|'):
        atelectasis_indexes.append(index)
        atelectasis_images.append(data_entry.iloc[index, 0])
    
    # Cardiomegaly
    elif 'Cardiomegaly' in i.split('|'):
        cardiomegaly_indexes.append(index)
        cardiomegaly_images.append(data_entry.iloc[index, 0])
        
    # Effusionv
    elif 'Effusion' in i.split('|'):
        effusion_indexes.append(index)
        effusion_images.append(data_entry.iloc[index, 0])   
    
    # Infiltration
    elif 'Infiltration' in i.split('|'):
        infiltration_indexes.append(index)
        infiltration_images.append(data_entry.iloc[index, 0]) 
    
    # Mass
    elif 'Mass' in i.split('|'):
        mass_indexes.append(index)
        mass_images.append(data_entry.iloc[index, 0]) 
    
    # Nodule
    elif 'Nodule' in i.split('|'):
        nodule_indexes.append(index)
        nodule_images.append(data_entry.iloc[index, 0]) 
    
    # Pneumonia
    elif 'Pneumonia' in i.split('|'):
        pneumonia_indexes.append(index)
        pneumonia_images.append(data_entry.iloc[index, 0])
    
    # Pneumothorax
    elif 'Pneumothorax' in i.split('|'):
        pneumothorax_indexes.append(index)
        pneumothorax_images.append(data_entry.iloc[index, 0]) 
```

```python
# Cheacking if a name is inside this column
empt = []
for i in data_entry.iloc[:, 1]: 
    if 'Normal' in i.split('|'):
        empt.append(i)
print(empt)
```
